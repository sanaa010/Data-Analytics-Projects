# TASK 3: Descriptive Statistics

### Calculate Mean, Median, Mode, Standard Deviation, and Variance for all Numerical Columns

```python
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
for col in numerical_columns:
    print(f"\nStatistics for {col}")
    print("Mean:", df[col].mean())
    print("Median:", df[col].median())
    print("Mode:", df[col].mode()[0])
    print("Standard Deviation:", df[col].std())
    print("Variance:", df[col].var())
```

### Findings
* Mean gives the average value of each numerical column.
* Median gives the middle value of the data.
* Mode shows the most frequently occurring value.
* Standard Deviation indicates how much the data is spread.
* Variance measures the variability in the data.

# Task 5: Cuisine Analysis (univariate and bivariate analysis)

```python
top_10_cuisines = df['Cuisines'].str.split(', ').explode().value_counts().head(10)

top_10_cuisines
```

#### 2. Visualize the Top 10 Most Common Cuisines using a Countplot

```python
cuisines = df['Cuisines'].str.split(', ').explode()

top_10 = cuisines.value_counts().head(10).index

plt.figure(figsize=(10,6))

sns.countplot(
    y=cuisines[cuisines.isin(top_10)],
    order=top_10,
    hue=cuisines[cuisines.isin(top_10)],
    palette='coolwarm',
    legend=False
)

plt.title('Top 10 Most Common Cuisines')
plt.xlabel('Number of Restaurants')
plt.ylabel('Cuisine')

plt.show()
```

# Task 6: Rating Analysis (univariate analysis)


#### 1. Distribution of Ratings
Definition

This analysis helps understand how restaurant ratings are distributed across the dataset and identifies the most common rating ranges.

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Aggregate rating'], bins=20, color='orchid')
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')

plt.show()
```

#### Findings
* Most restaurant ratings are concentrated between 2.5 and 4.0.
* A large number of restaurants have a rating of 0, indicating unrated restaurants.
* Very few restaurants have extremely low or extremely high ratings.

#### 2. Calculate Skewness of Ratings
Definition

Skewness measures the asymmetry of the rating distribution.

```python
skewness = df['Aggregate rating'].skew()

print("Skewness:", skewness)
```

#### Findings
* If the skewness value is negative, the distribution is left-skewed.
* If the skewness value is positive, the distribution is right-skewed.
* In this dataset, ratings are generally negatively skewed because many restaurants have moderate to high ratings.

#### 3. Detect Outliers Using IQR
Definition

The Interquartile Range (IQR) method is used to identify unusual rating values that differ significantly from the majority of the data.

```python
Q1 = df['Aggregate rating'].quantile(0.25)
Q3 = df['Aggregate rating'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['Aggregate rating'] < lower_bound) |
              (df['Aggregate rating'] > upper_bound)]

print("Number of Outliers:", len(outliers))
```

#### Findings
* The IQR method identifies ratings that fall outside the normal range.
* Most outliers are typically restaurants with rating 0 or unusually low ratings.
* Outliers can affect statistical analysis and should be examined carefully.

#### 4. Visualize Outliers Using Boxplot

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df['Aggregate rating'], color='lightskyblue')

plt.title('Boxplot of Aggregate Ratings')

plt.show()
```

#### Findings
* The boxplot shows that most restaurant ratings are concentrated between approximately 2.5 and 4.0.
* Several points appear outside the whiskers, indicating the presence of outliers in the dataset.
* Most of these outliers correspond to restaurants with very low ratings or unrated restaurants (rating = 0).
* The spread of ratings indicates variation in customer satisfaction across different restaurants.

# Task 10: Feature Engineering
#### 1. Create Cost Category

```python
def cost_category(cost):
    if cost <= 500:
        return 'Budget'
    elif cost <= 1500:
        return 'Moderate'
    elif cost <= 3000:
        return 'Premium'
    else:
        return 'Luxury'

df['Cost Category'] = df['Average Cost for two'].apply(cost_category)

df[['Average Cost for two','Cost Category']].head()
```

#### Analyze Cost Category

```python
df['Cost Category'].value_counts()
```

#### Visualization

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='Cost Category',
    hue='Cost Category',
    palette='viridis',
    legend=False
)

plt.title('Distribution of Cost Categories')
plt.show()
```

#### Findings
* Most restaurants belong to the Budget and Moderate categories.
* Premium restaurants are fewer in number.
* Luxury restaurants form the smallest group.
* The dataset is dominated by affordable and mid-range restaurants.

#### 2. Create Rating Category

```python
def rating_category(rating):
    if rating < 2.5:
        return 'Poor'
    elif rating < 3.5:
        return 'Average'
    elif rating < 4.5:
        return 'Good'
    else:
        return 'Excellent'

df['Rating Category'] = df['Aggregate rating'].apply(rating_category)

df[['Aggregate rating','Rating Category']].head()
```

#### Analyze Rating Category

```python
df['Rating Category'].value_counts()
```

#### Visualization

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='Rating Category',
    hue='Rating Category',
    palette='pastel',
    legend=False
)

plt.title('Distribution of Rating Categories')
plt.show()
```

#### Findings
* Most restaurants fall into the Good rating category.
* Fewer restaurants are classified as Poor or Excellent.
* Customer ratings are generally positive across the dataset.
* The majority of restaurants maintain satisfactory service quality.
